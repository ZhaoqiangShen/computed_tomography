```python
def load_and_show(filename, compare_to=None):  
    
    file = np.load(filename)
    
    print(filename)
    if compare_to is not None:
        mse = np.linalg.norm(file - compare_to)
        print("MSE: {}".format(mse))
    plt.figure(figsize=(8,8))
    plt.imshow(file, cmap='gray')
    plt.show()
    
    plt.imsave(filename[:-3]+'png', file, cmap='gray')
```
