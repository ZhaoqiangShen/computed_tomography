```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
full_reconstrction = np.load('full_reconstruction_1800_1800.npy')
tv_reconstruction_72_1800 = np.load('tv_reconstruction_72_1800.npy')
```

```python
def load_and_show(filename):  
    
    file = np.load(filename)
    mse = np.linalg.norm(file - full_reconstrction)
    print(filename)
    print("MSE: {}".format(mse))
    plt.figure(figsize=(8,8))
    plt.imshow(file, cmap='gray')
    plt.show()
```

```python
reconstructions = ['72_1800', '90_1800', '180_1800', '270_1800', '360_1800', '540_1800', '720_1800',  '900_1800']
```

```python
load_and_show('full_reconstruction_1800_1800.npy')
for run in reconstructions:
    tv_file = "tv_reconstruction_{}.npy".format(run)
    restricted_file = 'restricted_reconstruction_{}.npy'.format(run)
    load_and_show(tv_file)
    load_and_show(restricted_file)
    print("=======================")
```
